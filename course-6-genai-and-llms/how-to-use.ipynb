```python
from unsloth import FastLanguageModel

model, tokenizer = FastLanguageModel.from_pretrained(
    model_name="IRI2070/gemma-3-4b-it-unsloth-bnb-4bit-tax-quize-fine-tuned",
    max_seq_length=2048,
    load_in_4bit=True,
)
```

```python
system_prompt = """
#USE_TAGGED_OUTPUT#
شما یک دستیار هوش مصنوعی متخصص در پرسش‌های چهارگزینه‌ای مربوط به قانون مالیات ایران هستید.
وظیفه‌ی شما این است که سؤال، گزینه‌ها و زمینه‌ی ارائه‌شده را با دقت بخوانید، سپس گزینه‌ی صحیح
(A، B، C یا D) را انتخاب کرده و یک استدلال کوتاه اما روشن ارائه دهید.
راهنما:
- خروجی باید فقط شامل دو بخش باشد:
   <thinking> توضیح کوتاه به فارسی </thinking>
   <answer> یکی از A,B,C,D </answer>
- استدلال باید توضیح دهد چرا گزینه‌ی انتخاب‌شده درست است و چرا سایر گزینه‌ها نادرست هستند.
- لحن پاسخ باید حرفه‌ای، دقیق و معتبر باشد.
- در پاسخ، سؤال یا گزینه‌ها را دوباره بازنویسی نکنید.
"""

user_prompt = """
سوال: بر اساس ماده 218، مرجع تصویب آیین نامه مربوط به قسمت وصول مالیات و مرجع اجرا کدام اند؟

گزینه ها:
A: تصویب توسط وزارتخانه های امور اقتصادی و دارایی و دادگستری؛ اجرا توسط وزارت امور اقتصادی و دارایی
B: تصویب توسط مجلس؛ اجرا توسط قوه قضاییه
C: تصویب توسط سازمان امور مالیاتی؛ اجرا توسط خزانه داری کل کشور
D: تصویب توسط هیئت دولت؛ اجرا توسط وزارت دادگستری

متن زمینه:
# قانون مالیات های مستقیم 1366/12/03 ( با آخرین اصلاحات )

## ماده 215

تاریخ تصویب: 1366/12/03
مرجع تصویب: مجلس شورای اسلامی
تاریخ لازم الاجرا: 1368/01/01

ماده 215- در مورد اموال غير منقول توقيف شده در صورتي كه پس از دو نوبت آگهي ( كه نوبت دوم آن بدون حداقل قيمت آگهي مي
شود) خريداري براي آن پيدا نشود سازمان امور مالياتي كشور مي تواند مطابق ارزيابي كارشناس رسمي دادگستري معادل كل بدهي
مؤدي به علاوه هزينه متعلقه از مال مورد توقيف، تملك و بهاي آن را به حساب بدهي مؤدي منظور نمايد. تبصره 1- در صورتي كه
مؤدي قبل از انتقال مال مذكور به نام سازمان امور مالياتي كشور و يا ديگري، حاضر به پرداخت بدهي خود باشد سازمان امور
مالياتي كشور با دريافت بدهي مؤدي به اضافه ده درصد (10%) بدهي و هزينه هاي متعلقه از ملك مزبور رفع توقيف مي نمايد.
تبصره 2- در صورتي كه مال به تملك سازمان امور مالياتي كشور درآمده باشد اگر آمادگي جهت فروش داشته باشد و مؤدي درخواست
نمايد در شرايط مساوي اولويت با مؤدي است.

---

## ماده 218

تاریخ تصویب: 1366/12/03
مرجع تصویب: مجلس شورای اسلامی
تاریخ لازم الاجرا: 1368/01/01

ماده 218 – آئين نامه مربوط به قسمت وصول ماليات توسط وزارتخانه هاي امور اقتصادي و دارایي و دادگستري تصويب و توسط وزارت
امور اقتصادي و دارایي به موقع اجراء گذارده خواهد شد.

---

## نحوه اجرای ماده 215 قانون مالیات های مستقیم

نوع: بخشنامه
شماره: 200/1404/47
تاریخ صدور: 1404/05/05
مرجع تصویب: رئیس کل سازمان امور مالیاتی کشور

با عنایت به مفاد ماده 215 قانون مالیات های مستقیم، در مورد اموال غيرمنقول توقيف شده در صورتي كه پس از دو نوبت آگهي (
كه نوبت دوم آن بدون حداقل قيمت، آگهي مي شود) خريداري براي آن پيدا نشود، ادارات کل امور مالیاتی مي توانند مطابق
ارزيابي كارشناس رسمي دادگستري، معادل كل بدهي مؤدي به علاوه هزينه متعلقه را از مال مورد توقيف تملك نموده و بلافاصله پس
از تملک می بایست بهاي آن را از میزان بدهی مودی (اعم از اصل، جریمه و هزینه متعلق) کسر نمایند. همچنین ادارات کل پس از
تملک اموال مذکور مکلفند مطابق ضوابط قانونی، بدهی مالیاتی مودی (اعم از اصل و جریمه) را به حساب درآمدهای مالیاتی نزد
خزانه داری کل کشور واریز نموده یا در اسرع وقت، طی مکاتبه مستقیم با خزانه داری کل کشور، اموال مذکور را برای طی مراحل
قانونی معرفی نمایند تا توسط خزانه داری کل به حساب درآمدهای مالیاتی منظور گردد. تبصره: حفظ و نگهداری اموال موضوع این
ماده تا زمان تحویل به خزانه داری کل کشور بر عهده اداره کل ذی ربط خواهد بود.
"""

messages = [
    {
        'content': system_prompt,
        'role': 'system'
    },
    {
        'content': user_prompt,
        'role': 'user'
    }
]

prompt = tokenizer.apply_chat_template(messages, tokenize=False, add_generation_prompt=True).removeprefix('<bos>')

inputs = tokenizer(text=prompt, return_tensors="pt").to("cuda")

outputs = model.generate(**inputs, max_new_tokens=4_000, temperature=1, top_p=0.95, top_k=64)
response = tokenizer.decode(outputs[0], skip_special_tokens=True)

print(response)
```

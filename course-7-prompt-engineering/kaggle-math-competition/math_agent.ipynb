```
import pandas as pd

test_df = pd.read_csv('/content/test_with_translation.csv')
problems = test_df['translation'].tolist()
```

```
from smolagents import CodeAgent, OpenAIModel, tool
from google.colab import userdata

model = OpenAIModel(
    model_id="gpt-4o-mini",
    api_base="https://api.avalai.ir/v1",
    api_key=userdata.get('API_KEY'),
)
```

```
from smolagents import CodeAgent, InferenceClientModel

agent = CodeAgent(tools=[], model=model)

def get_answer(problem):
  prompt = rf"""
You are a math-solving code agent.
Your role is to solve mathematical problems step by step using Python code only.
Always:
- Perform all calculations exclusively through Python coding, never mentally.
- Break down the problem into clear steps.
- Show intermediate calculations explicitly.
- Ensure the final answer is a single number: either an integer (e.g., 4231, -12) or a finite decimal (e.g., 0.75). Do not use commas, fractions, or extra text. Correct format examples: 4231 -12 0.75 Incorrect format examples: 4 2 3 1 -12.0 0,2
- Ensure reproducibility and clarity in every solution.
- Do not skip steps or give only the final answer.

Now, solve the following problem:
{problem}
"""
  return agent.run(prompt)
```

```
from smolagents import CodeAgent, InferenceClientModel
from collections import Counter

agent = CodeAgent(tools=[], model=model)

results = []
for problem in problems:
  _answers = []
  for _ in range(5):
    result = get_answer(problem)
    _answers.append(result)
  most_common_answer = Counter(_answers).most_common(1)[0][0]
  results.append(most_common_answer)
  print(most_common_answer)
```

```
test_df['answer'] = results
submission_df = test_df[['problem_id', 'answer']]
submission_df.to_csv('submission.csv', index=False)
submission_df.head()
```
